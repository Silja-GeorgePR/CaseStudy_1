### 1. Read the dataset to the python environment.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("cars_data.csv")
data.head()
```

### 2. Check for the null values present in the dataset.


```python
print("count of null values present:")
print(data.isnull().sum())
```

### 3. Plot a bar graph of male vs female buyers participated in the sales.


```python
sns.countplot(data = data, x = 'Buyer Gender')
plt.title("Sales Participation based on gender")
plt.show()

```

### 4. Find the top 5 cars based on their sales price.


```python
df=data.nlargest(5, ['Sale Price'])
df[['Make','Model','Sale Price']]
```

### 5. Find the least 5 cars based on their Resell price.

```python
dff=data.nsmallest(5,['Resell Price'])
dff[['Make','Model','Resell Price']]
```
